# Data Munging

```python
import pandas as pd
```

```python
df = pd.read_csv('G:\Youtube\Pandas\example_files\Customers.csv')

customer_df = df

customer_df
```

### drop() 

```python
customer_df = customer_df.drop(columns='Profession', axis=1)

customer_df
```

```python
customer_df = customer_df.drop_duplicates()

customer_df
```

```python
pd.options.mode.chained_assignment = None
```

### strip()

```python
customer_df['Customer_name'] = customer_df.loc[:,'Customer_name'].str.strip('$#/*.')
customer_df
```

### replace()

```python
customer_df['Phone_no'] = customer_df.loc[:, 'Phone_no'].str.replace('[^A-Za-z0-9]','', regex=True)

customer_df
```

```python
customer_df['Gender'] = customer_df['Gender'].replace({
    'Male':'M',
    'm':'M',
    'MALE':'M',
    'male': 'M',
    'f': 'F',
    'female':'F',
    'FEMALE': 'F',
    'Female': 'F'

})

customer_df
```

```python
customer_df['Is_Non_vegetarian'] = customer_df['Is_Non_vegetarian'].replace({
    
    'Yes':'Y',
    'y':'Y',
    'YES':'Y',
    'yes':'Y',
    'No':'N',
    'NO': 'N',
    'n': 'N',
    'no':'N'
})

customer_df #Final dataframe after Handling Messy Data
```
